Nesse exercício, você deve criar um algoritmo que ordena uma lista de números. No entanto, a ordenação é feita da seguinte maneira:

1. Aplicando funções de sorteio (random), você deve embaralhar a lista.
2. Se após embaralhar, a lista estiver ordenada, o algoritmo encerra.
3. Se a lista não estiver ordenada, embaralha novamente.
4. Repita a partir do passo 1 até que a lista esteja ordenada.

OBS: Faça a **ordenação ser decrescente** (começa do maior valor e termina no menor valor)

```
lista_exemplo = [7, 3, 1, 4, 5]
```

```
import random
lista_atv1 = [7,3,1,4,5,9]
while(True):
  validador = True
  lista_ordenada = []
  pos_movidas = []
  anterior = lista_atv1[0]
  for i in range(1, len(lista_atv1)):
    if lista_atv1[i] > anterior: #se possição i da lista maior que o anterior sai do laço
      validador = False
      break
    anterior = lista_atv1[i] #se não anterior recebe a posição i da lista
  if validador: # se validador for verdadeiro sai do while
    break
  else:
    i = 0
    while i < len(lista_atv1):
       pos_escolhida = random.randint(0, len(lista_atv1)-1)
       nova_pos = True
       for j in range(0, len(pos_movidas)):
        if pos_escolhida == pos_movidas[j]: 
          nova_pos = False
          break
       if nova_pos:
        pos_movidas.append(pos_escolhida)
        lista_ordenada.append(lista_atv1[pos_escolhida])
        i+=1
  lista_atv1 = lista_ordenada
print(lista_atv1)
```

# 4. Faça manipulações em nível de pixel

Crie um algoritmo que receba 4 informações de um retângulo: posição x, posição y, altura e largura.

A seguir, faça com que uma nova imagem seja salva em disco, com um quadrado pintado com a cor preta ocupando as informações do retângulo.

Considere o exemplo abaixo:

image.png

```
import numpy as np
import cv2
from matplotlib import pyplot as plt
import matplotlib.patches as patches

entrada = input("Digite a posição X do retângulo: ")
posX = int(entrada)
entrada = input("Digite a posição Y do retângulo: ")
posY = int(entrada)
entrada = input("Digite a altura do retângulo: ")
alt = int(entrada)
entrada = input("Digite a largura do retângulo: ")
larg = int(entrada)

img = cv2.imread("/content/drive/My Drive/Colab Notebooks/PI/fundo.jpg")
fig,ax = plt.subplots(1)
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
ax.imshow(img)
rect = patches.Rectangle((posX,posY),alt,larg,edgecolor='black',facecolor='black')
ax.add_patch(rect)
plt.show()
```
